```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_total = len(df)
num_gp = df["school"].value_counts()["GP"]
num_ms = df["school"].value_counts()["MS"]

print("Total number of student records: ", num_total)
print("Percentage of students from Gabriel Pereira (GP) school: {0:.2%}".format(num_gp / num_total))
print("Percentage of students from Mousinho da Silveira (MS): {0:.2%}%".format(num_ms / num_total))
```

```python
num_total = len(df)
girls = df["sex"].value_counts()["F"]
boys = df["sex"].value_counts()["M"]

print("% girls in the study: {0:.2%}".format(girls / num_total))
print("% boys in the study: {0:.2%}".format(boys / num_total))
```

```python
gp_gender = df[df["school"] == "GP"]["sex"].value_counts()
ms_gender = df[df["school"] == "MS"]["sex"].value_counts()

schools_gender = pd.DataFrame([gp_gender, ms_gender])
schools_gender.index = ["GP", "MS"]
schools_gender.plot(kind="bar")
plt.title("Gender distribution by school")
plt.ylabel("No. of students")
plt.xlabel("Address type (F: female, M: male)")
```
